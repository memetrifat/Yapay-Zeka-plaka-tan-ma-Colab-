```python
import cv2
import imutils
import numpy as np
import pytesseract
from PIL import Image
from IPython.display import display, Image as IPImage

# Tesseract OCR yolu tanımlama
pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'

# Görüntüyü yükleme ve boyutlandırma
img = cv2.imread('araba3.jpg')
img = cv2.resize(img, (600,400))

# Gri tonlamalı ve kenar tespiti
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2.bilateralFilter(gray, 13, 15, 15)
edged = cv2.Canny(gray, 30, 200)

# Konturları bulma ve sıralama
contours = cv2.findContours(edged.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
contours = imutils.grab_contours(contours)
contours = sorted(contours, key = cv2.contourArea, reverse = True)[:10]
screenCnt = None

# Plaka konturunu bulma
for c in contours:
    peri = cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, 0.018 * peri, True)
    if len(approx) == 4:
        screenCnt = approx
        break

# Plaka konturunu işaretleyip plakayı kesme
if screenCnt is not None:
    cv2.drawContours(img, [screenCnt], -1, (0, 0, 255), 3)
    mask = np.zeros(gray.shape,np.uint8)
    new_image = cv2.drawContours(mask,[screenCnt],0,255,-1,)
    new_image = cv2.bitwise_and(img,img,mask=mask)
    (x, y) = np.where(mask == 255)
    (topx, topy) = (np.min(x), np.min(y))
    (bottomx, bottomy) = (np.max(x), np.max(y))
    Cropped = gray[topx:bottomx+1, topy:bottomy+1]
    # OCR ile plaka numarasını okuma
    text = pytesseract.image_to_string(Cropped, config='--psm 11')
    print("Plaka Tanıma Programlaması\n")
    print("Plaka Numarası:", text)

# Görüntüyü gösterme ve kaydetme
cv2.imwrite('arac.jpg', img)
cv2.imwrite('plaka.jpg', Cropped)

# Görüntüleri gösterme
display(IPImage('arac.jpg'))
display(IPImage('plaka.jpg'))
```
